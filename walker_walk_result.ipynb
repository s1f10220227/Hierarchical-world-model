```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('logdir/20250111-141144/metrics.jsonl', lines=True)
display(df.head(10))
display(df.tail(10))
```

```python
# 'episode/score' と 'episode/reward_rate' を抽出
scores = df['episode/score']
reward_rates = df['episode/reward_rate']
steps = df['step']
```

```python
# 可視化: スコアの推移
plt.figure(figsize=(10, 5))
plt.plot(steps, scores, label='Episode Score', color='b')
plt.xlabel('Step')
plt.ylabel('Score')
plt.title('Episode Score Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 可視化: リワードレートの推移
plt.figure(figsize=(10, 5))
plt.plot(steps, reward_rates, label='Reward Rate', color='g')
plt.xlabel('Step')
plt.ylabel('Reward Rate')
plt.title('Reward Rate Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
